* count - количество наблюдений (т.к. удалили пустые значения, поэтому одинаковые у всех переменных);
* mean - среднее значение переменной по имеющимся наблюдениям (x_{1}+...+x_{n}/n);
* std - стандартное отклонение значений, иными словами это корень разброса индивидуальных значений от ожидаемого, скорректированный на размер выборки 
* min - минимальное значение переменной
* 25% - если расположить значения выборки от минимума до максимума, то это значение, которое разделяет 25% наблюдений выборки (ближе к минимуму) и другие 75% наблюдений выборки
* 50% - медиана, показывает значение, которое отделяет первую половину выборки от второй
* 75% - аналогично 25%, только отделяет 75% наблюдений выборки (при счёте от минимума) от других 25% наблюдений выборки
* max - максимальное значение переменной

**10.** Добавьте в `df` столбец `fh_status`, в котором будут храниться типы стран в зависимости от значения индекса Freedom House (значения free, partly free, not free). Соответствие значений `fh09` с типами стран см. [здесь](https://freedomhouse.org/sites/default/files/2020-02/Methodology_FIW_2016.pdf) (в конце, Table 3).

```python
def indexing(fh):
    if 2.5 >= fh >= 1.0:
        return 'free'
    elif 5.0 >= fh >= 3.0:
        return 'partly free'
    elif 7.0 >= fh >= 5.5:
        return 'not free'

df.loc[:,'fh_status'] = df.loc[:, 'fh09'].apply(indexing)
df
```
