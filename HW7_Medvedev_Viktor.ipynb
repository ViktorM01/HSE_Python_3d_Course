### Задание 1

```python
import requests
from bs4 import BeautifulSoup

url = 'https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D0%B6%D0%B4%D1%83%D0%BD%D0%B0%D1%80%D0%BE%D0%B4%D0%BD%D1%8B%D0%B9_%D0%B8%D0%BD%D0%B4%D0%B5%D0%BA%D1%81_%D1%81%D1%87%D0%B0%D1%81%D1%82%D1%8C%D1%8F'
response = requests.get(url)

soup = BeautifulSoup(response.text)
print(soup.prettify())
```

```python
country_data = soup.find_all('tr', {'align':'center'})
print(soup.prettify())
```

```python
country_name = country_data[0].find('span', {'class' : 'nowrap'}).get('data-sort-value')
print(country_name)

HPI = country_data[0].find_all('td')[2].get_text()
print(HPI)

life_satisfaction = country_data[0].find_all('td')[3].get_text()
print(life_satisfaction)

life_span = country_data[0].find_all('td')[4].get_text()
print(life_span)

ecology = country_data[0].find_all('td')[5].get_text()[0:3]
print(ecology)
```

```python
info_per_country = [country_name, HPI, life_satisfaction, life_span, ecology]
info_per_country
```

```python
data_for_all = []

for country in country_data:
    country_name = country.find('span', {'class' : 'nowrap'}).get('data-sort-value')
    HPI = country.find_all('td')[2].get_text()
    life_satisfaction = country.find_all('td')[3].get_text()
    life_span = country.find_all('td')[4].get_text()
    ecology = country.find_all('td')[5].get_text()[0:3]
    
    list_country = [country_name, HPI, life_satisfaction, life_span, ecology]
    data_for_all.append(list_country)

print(data_for_all)
```

### Задание 2

```python
import pandas as pd
country_index_2012 = pd.DataFrame(data_for_all, columns = ['Страна', 'HPI', 'Удовлетворённость жизнью',
                                                      'Ожидаемая продолжительность жизни', 'Экологический след'])

country_index_2012 = country_index_2012.astype({'HPI' : 'float64', 'Удовлетворённость жизнью' : 'float64',
                                      'Ожидаемая продолжительность жизни' : 'float64', 'Экологический след' : 'float64'})
print(country_index_2012)
country_index_2012.describe()
```
